# Problem 1

```python
def prob1():
    """
    Create 3 visualizations of the crime data set.
    One of the visualizations should be a histogram.
    The visualizations should be clearly labelled and easy to understand.
    """
    # Load the Data
    df = pd.read_csv("crime_data.csv")
    
    # Create Plots
    df.plot(x="Year", y=["Murder","Forcible Rape", "Vehicle Theft"], title="Murder Rape and Vehicle Theft by Year", ylabel="Number of Crimes")
    df.plot(kind="scatter", x="Burglary", y="Robbery", title="Correlation between Burglary and Robbery Crimes in a Year")
    df[["Property", "Aggravated Assault","Larceny","Violent"]].hist(grid=False)
    plt.tight_layout()
    
 
```

# Problem 2

```python
 def prob2():
    """
    Use a visualization for each of the following to identify if trends exists 
    between Forcible Rape and
        1. Violent
        2. Burglary
        3. Aggravated Assault
    Plot each visualization.
    Return a tuple of booleans. Each entry should identify whether that 
    element correlates or not.
    """
    # Load the Data
    df = pd.read_csv("crime_data.csv")
    
    # Create Plots
    pd.plotting.scatter_matrix(df[['Forcible Rape', 'Violent']])
    plt.suptitle("Correlation between Rape and Violence")
    plt.tight_layout()
    pd.plotting.scatter_matrix(df[['Forcible Rape', 'Burglary']])
    plt.tight_layout()
    pd.plotting.scatter_matrix(df[['Forcible Rape', 'Aggravated Assault']])
    
    plt.tight_layout()
    
    return (True, False, True)
```

```python
print(prob2())
```

# Problem 3

```python
def prob3():
    """
    Use crime_data.csv to display the following distributions.
        1. The distributions of Burglary, Violent, and Vehicle Theft
        2. The distributions of Vehicle Theft against the values of Robbery
    """
    # Load the Data
    df = pd.read_csv("crime_data.csv")
    
    # Plot distributions
    df.plot(kind="box", y=['Burglary', 'Violent', 'Vehicle Theft'], vert=False, title="Distribution of Burglary, Violence and Vehicle Theft", xlabel="Number of Crimes", sharex=False)
    df.plot(kind='hexbin', x="Vehicle Theft", y="Robbery", gridsize=20, title="Vehicle Theft vs Robbery", xlabel="Vehicle Theft",sharex=False)
    
    
```

```python
prob3()
```

```python
def prob4():
    """
    Create 3 plots that compare variables or universities.
    These plots should answer questions about the data, 
    e.g. what is the distribution of graduation rates or do schools 
    with lower student to faculty ratios have higher tuition costs.
    
    These plots should be easy to understand, 
    have clear variable names, and citations.
    """
    # Load the Data
    df = pd.read_csv("college.csv")
    
    # First plot
    pd.plotting.scatter_matrix(df[['Grad.Rate', 'PhD']])
    plt.suptitle("Correlation between faculty with PHD and Graduation Rates")
    plt.tight_layout()
    plt.gcf().text(0, 0, "Figure 1: Source: US News and World Report's College Data https://cran.r-projects.org/web/packages/ISLR/ISLR.pdf")    
    
    # Second plot
    df.plot(kind='hexbin', x="Outstate", y="perc.alumni", gridsize=20, title="Alumni donations vs Out-of-state Tuition", sharex=False)
    plt.tight_layout()
    plt.gcf().text(0, 0, "Figure 2: Source: US News and World Report's College Data https://cran.r-projects.org/web/packages/ISLR/ISLR.pdf")    
    
    # Third plot
    df.plot(kind="box", y=['Apps', 'Accept', 'Enroll', 'Top25perc'], vert=False, title="Distribution of Applications, Acceptance, Enrollment, and Top Students")
    plt.gcf().text(0, 0, "Figure 3: Source: US News and World Report's College Data https://cran.r-projects.org/web/packages/ISLR/ISLR.pdf")
```
